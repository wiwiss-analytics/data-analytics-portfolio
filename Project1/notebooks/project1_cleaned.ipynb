```python
import pandas as pd

# Step 1: Load raw data
df = pd.read_csv("../raw/project1_raw_data.csv")

# Quick look
df.head()
df.shape

# How many duplicates?
print(df.duplicated().sum())

# Remove them
df = df.drop_duplicates()

# Check missing
df.isna().sum()
# Drop columns >30% missing
df = df.drop(columns=df.columns[df.isna().mean() > 0.3])
# Fill numeric columns with median
for col in df.select_dtypes(include="number").columns:
    df[col] = df[col].fillna(df[col].median())
# Fill text columns with "Unknown"
for col in df.select_dtypes(include="object").columns:
    df[col] = df# Make column names clean
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

# Example: convert kWh → MWh if exists
if "energy_kwh" in df.columns:
    df["energy_mwh"] = df["energy_kwh"] / 1000
    df = df.drop(columns=["energy_kwh"])


# Convert date columns
if "date" in df.columns:
    df["date"] = pd.to_datetime(df["date"], errors="coerce")

# Save cleaned CSV
df.to_csv("../cleaned/project1_cleaned.csv", index=False)


```
